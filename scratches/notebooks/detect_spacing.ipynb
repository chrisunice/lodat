```python
import numpy as np
from scipy import stats
```

```python
def infer_spacing(arr: np.ndarray) -> int | float:
    delta = np.diff(arr)
    mode_result = stats.mode(delta, keepdims=False)
    return mode_result.mode
```

```python
looks = np.arange(0, 360, 1)
infer_spacing(looks)
```

```python
deprs = np.arange(-90, 90, 5)
infer_spacing(deprs)
```
