```python
%matplotlib inline
import cv2
import lodat as lo
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(cv2_img: np.ndarray, title=None):
    fig, ax = plt.subplots()
    fig.set_size_inches(10, 10)
    ax.imshow(cv2.cvtColor(cv2_img, cv2.COLOR_BGR2RGB))
    ax.axis('off')
    if title is not None:
        ax.set_title(title)
    fig.show()
```

Load the image

```python
img_path = r"C:\LODAT\test_assets\xvv-311m000-f35a.X.baseupper.dat-003916-bwl2a.png"
img = cv2.imread(img_path)
plot(img, 'original')
```

Remove the wireframe

```python
naked_img, wireframe = lo.utils.extract_wireframe(img)
```

```python
plot(naked_img)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)
ax.scatter(wireframe[:, 0], wireframe[:, 1], c='white', s=1, edgecolors='black')
ax.invert_xaxis()
ax.axis('off')
fig.show()
```

```python
import plotly.graph_objects as go
import plotly.offline as po
# Read the PNG image using OpenCV
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Get image dimensions
height, width = img.shape

# Create a grid of x and y coordinates
x = np.arange(width)
y = np.arange(height)
x, y = np.meshgrid(x, y)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(x=x, y=y, z=img, colorscale='turbo')])

# Set axis labels
fig.update_layout(scene=dict(xaxis_title='X-axis (Pixel Position)', yaxis_title='Y-axis (Pixel Position)', zaxis_title='Intensity'))

# Show the plot
po.plot(fig)
```
